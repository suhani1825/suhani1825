**#  Practical to implement Data Manipulation using pandas Techniques.**

```python
import numpy as np
import pandas as pd
df=pd.DataFrame({"Rollno":[1,np.nan,3,4,5],
                 "Name":["sanika Tanaji Patil","Suhani Jagannath Saste","Mansi Sanjay Shelake","Omkar Jagannath Saste","Sakshi Sanjay She"],
                 "Marks":[50,80,90,31,35],
                 "Age":[25,7,2,10,np.nan],
                 "Class":["A","B",np.nan,"B","A"]})
df
```

```python
df.isnull().sum()
```

```python
df["rol"]=df["Rollno"].fillna(2)
df["rol"]
```

```python
mean_Age=df["Age"].mean()
mean_Age
```

```python
df["Age"]=df["Age"].fillna(mean_Age)
df["Age"]
```

```python
df
```

```python
median_Age=df["Age"].median()
median_Age
```

```python
df["Age"]=df["Age"].fillna(median_Age)
df
```

```python
mode_Age=df["Age"].mode()
mode_Age
```

```python
df["Age"]=df["Age"].fillna(mode_Age[3])
df
```

```python
df["First_name"]=df["Name"].str.split().str[1]
df
```
